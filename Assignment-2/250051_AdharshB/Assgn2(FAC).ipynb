```python
import numpy as np
import matplotlib.pyplot as plt
# initially im taking the input from all desired parameters
def plot_option_strategy():
    K = float(input("Enter Strike Price (K): "))
    P = float(input("Enter Premium (P): "))
    option_type = input("Enter Option Type (Call/Put): ").strip().capitalize()
    position = input("Enter Position (Buy/Sell): ").strip().capitalize()

    S = np.linspace(K - 50, K + 30, 80)
    # Based on the inputs the graphs will be plotted ( all 4 of em seperately )
    if option_type == "Call" and position == "Buy":
        profit = np.maximum(0, S - K) - P
        title, color = "Buying a Call Option", "green"
    elif option_type == "Call" and position == "Sell":
        profit = P - np.maximum(0, S - K)
        title, color = "Selling a Call Option", "red"
    elif option_type == "Put" and position == "Buy":
        profit = np.maximum(0, K - S) - P
        title, color = "Buying a Put Option", "blue"
    elif option_type == "Put" and position == "Sell":
        profit = P - np.maximum(0, K - S)
        title, color = "Selling a Put Option", "orange"
    else:
        print("Invalid input")
        return
    #plotted a graph of profit/loss vs spot prices
    plt.figure(figsize=(10, 6))
    plt.plot(S, profit, color=color, lw=2, label=f'{position} {option_type}')
    plt.axhline(0, color='black', linewidth=1)
    plt.title(title)
    plt.xlabel('Spot Price at Expiry (S)')
    plt.ylabel('Profit / Loss')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()


plot_option_strategy()
```
