# **Question 1**

```python
arr = np.arange(1,6,2)
n = len(arr)
for i in range(n):
    for j in range(i + 1, n):
      if arr[i] < arr[j]:
          arr[i], arr[j] = arr[j], arr[i]
print("max:", arr[0], "min:", arr[n-1])
```

# **Question 2**

```python
s = "aabbbbbaa"
state = 0
for i in range(len(s) // 2):
    if s[i] == s[len(s) - 1 - i]:
      continue
    else:
      print("not palindrome")
      state = 1
      break
if state ==0:
    print("palindrome")
```

# **Question 3**

```python
a = np.arange(1, 90)
x = 25
a.sort()
n = len(a)
i = 0
j = 1
while i < n and j < n:
  if a[j] - a[i] == x:
    print(a[i] + 1, a[j] + 1)
    i += 1
    j += 1
  elif a[j] - a[i] < x:
    j += 1
  else:
    i += 1
```
