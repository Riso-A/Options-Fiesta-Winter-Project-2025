```python
# Question 1

arr = list(map(int, input("Enter integers separated by space: ").split()))

min_val = arr[0]
max_val = arr[0]

for num in arr:
    if num < min_val:
        min_val = num
    if num > max_val:
        max_val = num

print("Minimum:", min_val)
print("Maximum:", max_val)
```

```python
# Question 2

s = input("Enter a string: ")

left = 0
right = len(s) - 1

while left < right:
    if s[left] != s[right]:
        print("Not a palindrome")
        break
    left += 1
    right -= 1
else:
    print("Palindrome")
```

```python
# Question 3

arr = list(map(int, input("Enter integers separated by space: ").split()))
x = int(input("Enter x: "))

arr.sort()

i = 0
j = 1

while j < len(arr):
    diff = arr[j] - arr[i]
    
    if diff == x:
        print("Pair found:", arr[i], arr[j])
        break
    elif diff < x:
        j += 1
    else:
        i += 1
```
